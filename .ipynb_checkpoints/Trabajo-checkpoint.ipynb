```python
import pandas as pd
import  matplotlib.pyplot as plt
```

```python
#1
#A
dfConfirmados = pd.read_csv('data/covid19_confirmados.csv', index_col="Country/Region")
#B
del dfConfirmados["Lat"]
del dfConfirmados["Long"]
#C
for i in range(1,461):
    dfConfirmados = dfConfirmados.rename(columns={dfConfirmados.columns[i]:pd.to_datetime(dfConfirmados.columns[i]).strftime('%d/%m/%Y') })

#D
del dfConfirmados["Province/State"]
dfConfirmados = dfConfirmados.groupby(by=['Country/Region']).sum()
dfConfirmados = dfConfirmados.sort_values('Country/Region')
display(dfConfirmados)
```

```python
dfPopulation = pd.read_csv('data/population_by_country_2020.csv', index_col="Country/Region")
dfPopulation = dfPopulation.sort_values('Country/Region')
display(dfPopulation)
```

```python
dfConfirmadosPopulation = pd.merge(dfConfirmados,dfPopulation[['Population (2020)']], on="Country/Region")
display(dfConfirmadosPopulation)
```

```python
dfConfirmadosPopulationUnion = pd.merge(dfConfirmados,dfPopulation[['Population (2020)']], on="Country/Region", how='outer')
noEstan = []
for i in range(0,len(dfConfirmadosPopulationUnion.index)):
    if not (dfConfirmadosPopulationUnion.index[i] in dfConfirmadosPopulation.index):
        noEstan.append(dfConfirmadosPopulationUnion.index[i])

print('En las siguientes filas no se pudo realizar merge: ')
print(noEstan)
```

```python
#1
#A
dfRecuperados = pd.read_csv('data/covid19_recuperados.csv', index_col="Country/Region")
#B
del dfRecuperados["Lat"]
del dfRecuperados["Long"]
#C
for i in range(1,461):
    dfRecuperados = dfRecuperados.rename(columns={dfRecuperados.columns[i]:pd.to_datetime(dfRecuperados.columns[i]).strftime('%d/%m/%Y') })
    
#D
del dfRecuperados["Province/State"]
dfRecuperados = dfRecuperados.groupby(by=['Country/Region']).sum()
dfRecuperados = dfRecuperados.sort_values('Country/Region')
display(dfRecuperados)


dfRecuperadosPopulation = pd.merge(dfRecuperados,dfPopulation[['Population (2020)']], on="Country/Region")
display(dfRecuperadosPopulation)

dfRecuperadosPopulationUnion = pd.merge(dfRecuperados,dfPopulation[['Population (2020)']], on="Country/Region", how='outer')
noEstan = []
for i in range(0,len(dfRecuperadosPopulationUnion.index)):
    if not (dfRecuperadosPopulationUnion.index[i] in dfRecuperadosPopulation.index):
        noEstan.append(dfRecuperadosPopulationUnion.index[i])

print('En las siguientes filas no se pudo realizar merge: ')
print(noEstan)
```

```python
#1
#A
dfDecesos = pd.read_csv('data/covid19_decesos.csv', index_col="Country/Region")
#B
del dfDecesos["Lat"]
del dfDecesos["Long"]
#C
for i in range(1,461):
    dfDecesos = dfDecesos.rename(columns={dfDecesos.columns[i]:pd.to_datetime(dfDecesos.columns[i]).strftime('%d/%m/%Y') })
    
#D
del dfDecesos["Province/State"]
dfDecesos = dfDecesos.groupby(by=['Country/Region']).sum()
dfDecesos = dfDecesos.sort_values('Country/Region')
display(dfDecesos)


dfDecesosPopulation = pd.merge(dfDecesos,dfPopulation[['Population (2020)']], on="Country/Region")
display(dfDecesosPopulation)

dfDecesosPopulationUnion = pd.merge(dfDecesos,dfPopulation[['Population (2020)']], on="Country/Region", how='outer')
noEstan = []
for i in range(0,len(dfDecesosPopulationUnion.index)):
    if not (dfDecesosPopulationUnion.index[i] in dfDecesosPopulation.index):
        noEstan.append(dfDecesosPopulationUnion.index[i])

print('En las siguientes filas no se pudo realizar merge: ')
print(noEstan)
```

### Análisis global (1.5pt)

1. Obtenga la cantidad total  a la fecha (26 de Abril) de los casos (a) confirmados, (b) cerrados (recuperados más decesos) y (c) activos (confirmados menos cerrados) por país. En base a estas cantidades calcule el porcentaje de 
    1. casos activos con respecto al total de confirmados
    1. decesos con respecto al total de casos cerrados
    1. recuperados con respecto al total de casos cerrados
1. Luego calcule las siguientes tasas en base a información a la fecha
    1. Tasa de incidencia: confirmados por 100.000 dividido población total
    1. Tasa de recuperación: recuperados por 100.000 dividido población total
    1. Tasa de mortalidad: decesos por 100.000 dividido población total
1. En base a lo anterior muestre los nombres y los valores de los 5 países con
    1. La mayor y menor tasa de incidencia, respectivamente
    1. La mayor y menor tasa de recuperación, respectivamente
    1. La mayor y menor tasa de mortalidad, respectivamente
1. Responda, analice y discuta: ¿En qué lugar/ranking se encuentra Chile con respecto a cada una de las métricas de los puntos anteriores? ¿A qué países de la [OCDE](https://es.wikipedia.org/wiki/Organizaci%C3%B3n_para_la_Cooperaci%C3%B3n_y_el_Desarrollo_Econ%C3%B3micos) se parece más? ¿A qué paises de latinoamérica se parece más? 

```python
NumeroConfirmadosTotal = dfConfirmadosPopulation[pd.to_datetime('2021-04-25').strftime('%d/%m/%Y')]
NumeroCerradosTotal = dfDecesosPopulation[pd.to_datetime('2021-04-25').strftime('%d/%m/%Y')] + dfRecuperadosPopulation[pd.to_datetime('2021-04-25').strftime('%d/%m/%Y')]
NumeroActivosTotal = dfConfirmadosPopulation[pd.to_datetime('2021-04-25').strftime('%d/%m/%Y')] - NumeroCerradosTotal
```

```python
#A
print('Casos activos con respecto al total de confirmados \n')
ActivosRespectoConfirmados = (NumeroActivosTotal * 100) / NumeroConfirmadosTotal
display(ActivosRespectoConfirmados)
print('\n')

#B
print('Decesos con respecto al total de casos cerrados\n')
DecesosRespectoCerrados = (dfDecesos[pd.to_datetime('2021-04-25').strftime('%d/%m/%Y')] * 100) / NumeroCerradosTotal
display(DecesosRespectoCerrados)
print('\n')

#C
print('Recuperados con respecto al total de casos cerrados\n')
RecuperadosRespectoCerrados = (dfRecuperados[pd.to_datetime('2021-04-25').strftime('%d/%m/%Y')] * 100) / NumeroCerradosTotal
display(RecuperadosRespectoCerrados)
print('\n')
```

```python
#2
#A
print('Tasa de incidencia \n')
TasaIncidencia = (NumeroConfirmadosTotal * 100000) / dfConfirmadosPopulation['Population (2020)']
display(TasaIncidencia)
print('\n')

#B
print('Tasa de recuperacion \n')
TasaRecuperacion = (dfRecuperadosPopulation[pd.to_datetime('2021-04-25').strftime('%d/%m/%Y')] * 100000) / dfRecuperadosPopulation['Population (2020)']
display(TasaRecuperacion)
print('\n')

#C
print('Tasa de Mortalidad \n')
TasaMortalidad = (dfDecesosPopulation[pd.to_datetime('2021-04-25').strftime('%d/%m/%Y')] * 100000) / dfDecesosPopulation['Population (2020)']
display(TasaMortalidad)
```

```python
#3

#A
print('Mayor tasa de incidencia \n')
MayorTasaIncidencia = TasaIncidencia.sort_values(ascending=False)[0:5]
display(MayorTasaIncidencia)
print('\n')
print('Menor tasa de incidencia \n')
MenorTasaIncidencia = TasaIncidencia.sort_values()[0:5]
display(MenorTasaIncidencia)
print('\n')

#B
print('Mayor tasa de recuperacion \n')
MayorTasaRecuperacion = TasaRecuperacion.sort_values(ascending=False)[0:5]
display(MayorTasaRecuperacion)
print('\n')
print('Menor tasa de recuperacion \n')
MenorTasaRecuperacion = TasaRecuperacion.sort_values()[0:5]
display(MenorTasaRecuperacion)
print('\n')

#C
print('Mayor tasa de mortalidad \n')
MayorTasaMortalidad = TasaMortalidad.sort_values(ascending=False)[0:5]
display(MayorTasaMortalidad)
print('\n')
print('Menor tasa de mortalidad \n')
MenorTasaMortalidad = TasaMortalidad.sort_values()[0:5]
display(MenorTasaMortalidad)
print('\n')
```
