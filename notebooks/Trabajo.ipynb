### Preparación de datos 

##### 1. Explore el archivo `covid19_confirmados.csv` y configure la función `read_csv` de pandas para importarlo como un `DataFrame`. Utilice el nombre de país como índice principal

```python
dfConfirmados = pd.read_csv('data/covid19_confirmados.csv', index_col="Country/Region")
# Importamos el DataFrame 'covid19_confirmados.csv' y definimos 
# el nombre del paìs como ìndice principal.
```

#####  Elimine las columnas `Lat` y `Long`

```python
del dfConfirmados["Lat"]
del dfConfirmados["Long"]
```

##### Use TimeStamps para los "nombres de columna" asociados a fechas

```python
for i in range(1,461):
    dfConfirmados = dfConfirmados.rename(columns={dfConfirmados.columns[i]:pd.to_datetime(dfConfirmados.columns[i]).strftime('%d/%m/%Y') })
# Renombramos las columnas por las fechas
```

##### Reduzca la columna `Province/State`, el DataFrame resultante debe tener los totales a nivel país

```python
# Reducimos las columna 'Province/State', teniendo el total
# a nivel paìs.
del dfConfirmados["Province/State"]
dfConfirmados = dfConfirmados.groupby(by=['Country/Region']).sum()
dfConfirmados = dfConfirmados.sort_values('Country/Region')
```

##### Mostramos el DataFrame 

```python
display(dfConfirmados)
```

##### Explore el archivo `population_by_country_2020.csv` y configure la función `read_csv` de pandas para importarlo como un `DataFrame`. 

```python
dfPopulation = pd.read_csv('data/population_by_country_2020.csv', index_col="Country/Region")
dfPopulation = dfPopulation.sort_values('Country/Region')
# Importamos el DataFrame 'Population', definiendo el nombre del 
# paìs como ìndice principal.
```

##### Mostramos el DataFrame 

```python
display(dfPopulation)
```

```python
# Creamos un nuevo DataFrame a partir de hacer merge
# entre los dos DataFrames anteriores.
dfConfirmadosPopulation = pd.merge(dfConfirmados,dfPopulation[['Population (2020)']], on="Country/Region")
display(dfConfirmadosPopulation)
```

#####  Haga un `merge` de los `DataFrame` anteriores. El objetivo es asignar un valor de `Population` a cada elemento del primer `DataFrame`. Indique las filas donde no se puede realizar el `merge` y luego descártelas de su `DataFrame` final

```python
# Indicamos las filas en las que no se pudieron hacer merge.
dfConfirmadosPopulationUnion = pd.merge(dfConfirmados,dfPopulation[['Population (2020)']], on="Country/Region", how='outer')
noEstan = []
for i in range(0,len(dfConfirmadosPopulationUnion.index)):
    if not (dfConfirmadosPopulationUnion.index[i] in dfConfirmadosPopulation.index):
        noEstan.append(dfConfirmadosPopulationUnion.index[i])
```
