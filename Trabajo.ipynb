```python
import pandas as pd
import  matplotlib.pyplot as plt
```

```python
#1
#A
dfConfirmados = pd.read_csv('data/covid19_confirmados.csv', index_col="Country/Region")
#B
del dfConfirmados["Lat"]
del dfConfirmados["Long"]
#C
for i in range(1,461):
    dfConfirmados = dfConfirmados.rename(columns={dfConfirmados.columns[i]:pd.to_datetime(dfConfirmados.columns[i]) })

#D
del dfConfirmados["Province/State"]
dfConfirmados = dfConfirmados.groupby(by=['Country/Region']).sum()
dfConfirmados = dfConfirmados.sort_values('Country/Region')
display(dfConfirmados)
```

```python
dfPopulation = pd.read_csv('data/population_by_country_2020.csv', index_col="Country/Region")
dfPopulation = dfPopulation.sort_values('Country/Region')
display(dfPopulation)
```

```python
dfConfirmadosPopulation = pd.merge(dfConfirmados,dfPopulation[['Population (2020)']], on="Country/Region")
display(dfConfirmadosPopulation)
```

```python
dfConfirmadosPopulationUnion = pd.merge(dfConfirmados,dfPopulation[['Population (2020)']], on="Country/Region", how='outer')
noEstan = []
for i in range(0,len(dfConfirmadosPopulationUnion.index)):
    if not (dfConfirmadosPopulationUnion.index[i] in dfConfirmadosPopulation.index):
        noEstan.append(dfConfirmadosPopulationUnion.index[i])

print('En las siguientes filas no se pudo realizar merge: ')
print(noEstan)
```

```python
#1
#A
dfRecuperados = pd.read_csv('data/covid19_recuperados.csv', index_col="Country/Region")
#B
del dfRecuperados["Lat"]
del dfRecuperados["Long"]
#C
for i in range(1,461):
    dfRecuperados = dfRecuperados.rename(columns={dfRecuperados.columns[i]:pd.to_datetime(dfRecuperados.columns[i]) })
    
#D
del dfRecuperados["Province/State"]
dfRecuperados = dfRecuperados.groupby(by=['Country/Region']).sum()
dfRecuperados = dfRecuperados.sort_values('Country/Region')
display(dfRecuperados)


dfRecuperadosPopulation = pd.merge(dfRecuperados,dfPopulation[['Population (2020)']], on="Country/Region")
display(dfRecuperadosPopulation)

dfRecuperadosPopulationUnion = pd.merge(dfRecuperados,dfPopulation[['Population (2020)']], on="Country/Region", how='outer')
noEstan = []
for i in range(0,len(dfRecuperadosPopulationUnion.index)):
    if not (dfRecuperadosPopulationUnion.index[i] in dfRecuperadosPopulation.index):
        noEstan.append(dfRecuperadosPopulationUnion.index[i])

print('En las siguientes filas no se pudo realizar merge: ')
print(noEstan)
```
